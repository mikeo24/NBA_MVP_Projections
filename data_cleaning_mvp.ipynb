## Cleaning the players data

```python
players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
#Some players have multiple rows within a season because they probably got traded multiple times

players.groupby(["Player","Year"]).get_group(("Greg Anderson", 1991))
```

```python
#Dealing duplicate rows

def single_row(df):
    if df.shape[0]==1:    #If there's only one row for that specific player return the data
        return df
    else:
        row = df[df["Tm"] == "TOT"]       #If they have multiple teams they payed for just return their total stats (TOT)
        row["Tm"] = df.iloc[-1,:]["Tm"]   #replace TOT with the last team that player played for in the last row
        return row
    
players = players.groupby(["Player", "Year"]).apply(single_row)
```

## Clean the team data

```python
teams = pd.read_csv("teams.csv")
```

```python
teams.head(30)
```

```python
#Removing every rows that have division in it (~ means NOT) 

teams = teams[~teams["W"].str.contains("Division")] # If the string has division in it get rid of the entire row
```

```python
teams.head(10)
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
teams
```

```python
teams["Team"].unique()
```

```python
combined["Tm"].unique()
```

```python
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        ce, abbrev,name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
    
```

```python
nicknames
```

```python
#We can now use map to create a new column to find the Team name based on the column TM

combined["Team"] = combined["Tm"].map(nicknames)
```

## Data Exploration

```python
# Top 5 people with the most points
# Anyone who scored the most the points in a season


highest_scoring = stats[stats["G"] > 70].sort_values("PTS", ascending=False).head(5)
highest_scoring
```

```python
highest_scoring.plot.bar("Player", "PTS")
```

```python
# who scored the highest each year
# Lambda function is just the equivalent of a regular function in one line 

highest_scoring = stats.groupby("Year").apply(lambda x: x.sort_values("PTS", ascending=False).head(1))
highest_scoring
# we sort by PTS and have the highest at the top and using head(1) to get the highest since it is on top
```

```python
highest_rebds = stats.groupby("Year").apply(lambda x: x.sort_values("AST", ascending=False).head(1))
highest_rebds
```
