## Training a machine Learning model

```python
stats.columns
```

```python
# For our predictors we remove every categorical variables and the variable we want to predict ("Share") 
#So we are removing everything that directly correlate to share (Pts Won a& Pts Max)

predictors = [ 'Age', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P',
'3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV','PF', 'PTS', 'Year', 'W', 'L', 'W/L%', 'GB', 'PS/G',
'PA/G', 'SRS']

```

```python
# for Train we are going to train everything before 2021

train = stats[stats["Year"] < 2022]
train.columns
```

```python
#predict our train data on the test data
#Not to test on data that is before data you are training
# You cannot train the data through 1995 to 2022 and then make prediction on 1994 because that's info the algorithm  
#wouldn't have if it was running in the real world
# when having time series data be careful not to use data from after your test
test = stats[stats["Year"] == 2022]
```

```python
# Ridge regression is degsign to prevent overfitting, it shrinks the linear regression coefficient 

from sklearn.linear_model import Ridge

reg = Ridge(alpha=.1) # alpha just control how much the coefficient is going to be shrunk to prevent overfitting
```

```python
reg.fit(train[predictors], train["Share"])
```

```python
predictions =reg.predict(test[predictors])
```

```python
predictions = pd.DataFrame(predictions, columns = ["predictions"], index=test.index)
predictions
```

```python
# We are printing now the share of every player in the 2022 of winning MVP 

combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
```

```python
# Out of the top 5 how many did we actually predict the players in the top 5
```

```python
def find_ap(combination):
    actual = combination.sort_values("Share", ascending= False).head(5)
    predicted = combination.sort_values("predictions", ascending= False)
    ps = []
    found= 0
    seen = 1
    for index, row in predicted.iterrows():
        if row["Player"] in actual["Player"].values:
            found += 1
            ps.append(found/seen)
        seen += 1
    return sum(ps) / len(ps)
```

```python
find_ap(combination)

# This is a good error metric and tells us how accurate our model predict the top 5 MVPs
```

## implementing backtesting to predict each year

```python
years = list(range(1991,2023))
```

```python
# It will help us make predictions for every year from 1996 to 2022


aps = []
all_predictions =[] 
for year in years[6:]:      # years[5:] = 1996
    train=stats[stats["Year"] < year]   # predict every year starting in 1996
    test = stats[stats["Year"] == year]   # test data will always be the year we are trying to predict
    reg.fit(train[predictors], train["Share"])
    predictions = reg.predict(test[predictors])
    predictions = pd.DataFrame(predictions, columns = ["predictions"], index=test.index)
    combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
    all_predictions.append(combination)
    aps.append(find_ap(combination))
```

```python
sum(aps) / len(aps)

# based on our alghorithm our predictions is 70% accurate 
```

```python
def add_ranks(combination):
    combination = combination.sort_values("Share", ascending= False)
    combination["Rk"] = list(range(1, combination.shape[0]+1))
    combination = combination.sort_values("predictions", ascending= False)
    combination["predicted_Rk"] = list(range(1, combination.shape[0]+1))
    combination["Diff"] = combination["Rk"] - combination["predicted_Rk"]
    return combination
```

```python
# This our prediction for the 1998 season. It gives us the actual rank, predicted rank and Difference rank
# we sort it by the people who had the biggest difference Diff
ranking = add_ranks(all_predictions[1])
ranking[ranking["Rk"] < 6].sort_values("Diff", ascending = False)
```

```python
def backtest(stats, model, year, predictors):
    aps = []
    all_predictions =[] 
    for year in years[6:]:      # years[5:] = 1996
        train=stats[stats["Year"] < year]   # predict every year starting in 1996
        test = stats[stats["Year"] == year]   # test data will always be the year we are trying to predict
        model.fit(train[predictors], train["Share"])
        predictions = reg.predict(test[predictors])
        predictions = pd.DataFrame(predictions, columns = ["predictions"], index=test.index)
        combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
        combination = add_ranks(combination)
        all_predictions.append(combination)
        aps.append(find_ap(combination))
        
    return sum(aps)/len(aps), aps, pd.concat(all_predictions)
```

```python
mean_ap, apps, all_predictions = backtest(stats, reg, years[5:], predictors)
```

```python
mean_ap
```

## Adding more predictors

```python
# one way to add more predictors to our model to make it more accurate
# we can look at the ratio between the season averages for all the players and look at a particular player
# for example if you tripled the scoring average in a season that might tell the alghorithm something

#The mean in our lambda function will calculate the mean of PTS, AST, STL, ETC.. and then divide every player average 
# by the mean --- if a player number is below the mean then you will have .1, .2 --- if the player number is above the
# mean you will have 2, 3 , 4 as value

stat_ratios = stats[["PTS", "AST", "STL", "BLK", "3P", "Year"]].groupby("Year").apply(lambda x: x/x.mean()) 
```

```python
stat_ratios
```

```python
stats[["PTS_T", "AST_R", "STL_R", "BLK_R", "3P_R"]] = stat_ratios[["PTS", "AST", "STL", "BLK", "3P"]]
```

```python
predictors += ["PTS_T", "AST_R", "STL_R", "BLK_R", "3P_R"]
```

```python
mean_ap, apps, all_predictions = backtest(stats, reg, years[5:], predictors)
```
